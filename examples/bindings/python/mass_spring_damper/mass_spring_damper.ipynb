```python
from vanta_core_py.ode import euler_forward, euler_backward, runge_kutta_4
import matplotlib.pyplot as plt
```

```python
# Variables
k = 0.2  # Spring constant
c = 0.2  # Damping coefficient
t0 = 0.0
t1 = 100.0
y0 = [1.0, 0.0]
h = 0.1
```

```python
# Mass spring damper system
def mass_spring_damper(t, y):
    dydt = [y[1], -c * y[1] - k * y[0]]
    return dydt
```

```python
# Compute solutions using different numerical methods
euler_forward_solution = euler_forward(mass_spring_damper, t0, t1, y0, h)
euler_backward_solution = euler_backward(mass_spring_damper, t0, t1, y0, h)
runge_kutta_4_solution = runge_kutta_4(mass_spring_damper, t0, t1, y0, h)
```

```python
%matplotlib widget

# Plot
plt.figure()
plt.plot(
    euler_forward_solution.t, euler_forward_solution.y[:, 0], label="Euler Forward"
)
plt.plot(
    runge_kutta_4_solution.t, runge_kutta_4_solution.y[:, 0], label="Runge Kutta 4"
)
plt.plot(
    euler_backward_solution.t, euler_backward_solution.y[:, 0], label="Euler Backward"
)
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("Distance (m)")
plt.legend()
plt.show()
```
